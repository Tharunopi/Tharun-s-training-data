```python
import pandas as pd
import numpy as np
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer, KNNImputer
```

```python
df = pd.read_csv(r"E:\Zip files\playground-series-s4e12\train.csv", dtype={"Policy Start Date":str})
df.head()
```

```python
df = df.drop(['id'], axis=1)
```

```python
df['Policy Start Date'] = df['Policy Start Date'].apply(lambda x: pd.to_datetime(x).timestamp())
```

```python
col_to_transform = ["Gender", "Marital Status", "Education Level", "Occupation", "Location", "Policy Type", "Customer Feedback", "Smoking Status", "Exercise Frequency", "Property Type"]
df_imp_mean = ["Age", "Annual Income", "Number of Dependents", "Health Score", "Vehicle Age", "Credit Score", "Insurance Duration"]
df_imp_repeated = ["Marital Status", "Occupation", "Previous Claims", "Customer Feedback"]
imp_mean = [0, 2, 4, 7, 11, 12, 13]
imp_repated_values = [3, 6, 10, 15]
```

```python
for i in range(len(df_imp_mean)):
    imp = SimpleImputer(
        missing_values=np.nan, 
        strategy="mean"
    )
    df[df_imp_mean[i]] = imp.fit_transform(df.iloc[:, imp_mean[i]].values.reshape(-1, 1))[:, 0]
```

```python
for i in range(len(df_imp_repeated)):
    imp = SimpleImputer(
        missing_values=np.nan, 
        strategy='most_frequent'
    )
    df[df_imp_repeated[i]] = imp.fit_transform(df.iloc[:, imp_repated_values[i]].values.reshape(-1, 1))[:, 0]
```
