```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import math
```

```python
df = pd.read_csv("E:\BlackClover\Machine Learning-A-Z-Codes-Datasets (1)\Machine Learning A-Z (Codes and Datasets)\Part 6 - Reinforcement Learning\Section 32 - Upper Confidence Bound (UCB)\Python\Ads_CTR_Optimisation.csv")
df.shape
```

```python
N =1000
d = df.shape[1] 
ads_selected = []
num_ads_selected = [0] * d
sum_total_rewards = [0] * d 
total_reward = 0
for i in range(0, N):
    ad = 0
    max = 0
    for j in range(0, d):
        if num_ads_selected[j] > 0:
            average = sum_total_rewards[j] / num_ads_selected[j]
            delta = math.sqrt(3/2 * math.log(i+1) / num_ads_selected[j])
            ucb = average + delta
        else:
            ucb = 1e400
        if ucb > max:
            max = ucb
            ad = j
    ads_selected.append(ad)
    num_ads_selected[ad] += 1
    reward = df.values[i, ad]
    sum_total_rewards[ad] += reward
    total_reward += reward
```
